```python
url= "http://www.iso.org/iso/annual_report_2009.pdf"
r = requests.get(url, stream = True)
chunk_size = 256
total_size = int(r.headers['Content-Length'])
with open("sample.pdf", "wb") as file:
    bytes_written = 0
    for chunk in r.iter_content(chunk_size):
            bytes_written += file.write(chunk)
            percent = (bytes_written/total_size)*100
            print(percent)
```

```python
from tqdm import tqdm
url= "http://www.iso.org/iso/annual_report_2009.pdf"
r = requests.get(url, stream = True)
chunk_size = 256
total_size = int(r.headers['Content-Length']) // 256 #no of iterations we need to perform to completely download the file.

with open("sample1.pdf", "wb") as file:
    bytes_written = 0
    for chunk in tqdm(r.iter_content(chunk_size), total = total_size):
            file.write(chunk)
```

# Web Scraping

```python
from bs4 import BeautifulSoup
```

```python
url = "https://www.snapdeal.com/search"
params = {
    "keyword": "iphone xs"
}
r=requests.get(url, params = params)
```

```python
r.url
```

```python
print(r.content.decode('utf-8'))
```

```python
soup = BeautifulSoup(r.content) #parses the content i.e., html document in tree structure
```

```python
divs = soup.findAll('div', attrs = {"class": "product-tuple-listing"})
```

```python
divs
```

```python
for div in divs:
    etitle = div.find('p', attrs = {"class": "product-title"})
    eprice = div.find('div', attrs = {"class": "product-price-row"})
    eimage = div.find('picture', attrs = {"class": "picture-elem"})
    epic = eimage.find('img', attrs={"class":"product-image"}).attrs['src']
    print(etitle.text)
    print(eprice.text)
    print(epic)
```

```python
url = "https://www.pepperfry.com/site_product/search"
headers = {"User-Agent" :"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/74.0.3729.169 Chrome/74.0.3729.169 Safari/537.36"}
params = {
    "q": "bean bags"
}
r=requests.get(url, params=params, headers = headers)
```

```python
r.url
```

```python
r.headers
```

```python
type(r.content.decode('utf-8'))
```

```python
soup = BeautifulSoup(r.content)
```

```python
soup
```

```python
divs = soup.findAll('div', attrs = {"class": "srch-rslt-cards"})
```

```python
divs
```

```python
li=[]
```

```python
for div in divs:
    dictionary = {}
    Title = div.find('div', attrs={"class": "clip-dtl-ttl"}).a
    Price = div.find('span', attrs={"class": "clip-offr-price"})
    DiscountPrice = div.find('span', attrs={"class": "clip-ofr-save-per"})
    RetailPrice = div.find('span', attrs={"class": "clip-retail-price"})
    image = div.find('img').attrs['data-src']

    dictionary["Title"] = Title.text.strip()
    dictionary["Price"] = Price.text
    dictionary["DiscountPrice"] = DiscountPrice.text
    dictionary["RetailPrice"] = RetailPrice.text
    dictionary["Image"] = image
    
li.append(dictionary)

```

```python
li
```

```python
with open("pepperfry.json", "w+") as file:
    json.dump(li, file)
```
