```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

baseUrl = ("https://www.nseindia.com/live_market/dynaContent/live_watch/option_chain/optionKeys.jsp?symbolCode=-10006&symbol=NIFTY&symbol=NIFTY&instrument=-&date=-&segmentLink=17&symbolCount=2&segmentLink=17")
```

```python
page = requests.get(baseUrl)
```

```python
soup = BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
table_it = soup.find_all(class_ = "opttbldata")
table_cls_1 = soup.find_all(id="octable")
```

```python
print(table_cls_1)
```

```python
col_list = []

for mytable in table_cls_1:
    tableHead = mytable.find('thead')
    
    try:
        rows = tableHead.find_all('tr')
        for tr in rows:
            cols = tr.find_all('th')
            for th in cols:
                er = th.text
                ee = er.encode('utf8')
                col_list.append(ee)
    except:
        print("no thread")
        
col_list_fnl = [e for e in col_list if e not in ('CALLS', 'PUTS','Charts','\xc2\xa0' )]
```

```python
table_cls_2 = soup.find(id="octable")
all_trs = table_cls_2.find_all('tr')
req_row = table_cls_2.find_all('tr')
```

```python
table_cls_2
```

```python
req_row
```

```python
new_table = pd.DataFrame(index=range(0,len(req_row)-3), columns = col_list_fnl)
row_marker = 0
for rowNo, trNos in enumerate(req_row):
    if rowNo <=1 or rowNo == len(req_row)-1:
        continue
    
    td_columns = trNos.find_all('td')
    
    select_cols = td_columns[1:22]
    cols_horizontal = range(0,len(select_cols))
    
    for nu, column in enumerate(select_cols):
        utf_string = column.get_text()
        utf_string = utf_string.strip('\n\r\t": ')
        tr = utf_string.encode('utf8')
        tr = tr.replace(','.encode('utf-8'), ''.encode('utf8'))
        new_table.iloc[row_marker,[nu]] = tr
        
    row_marker += 1
    
print (new_table)
new_table.to_csv('Chain_Table1.csv')
```

```python
tableBody = [] #important
for tb in table.findAll('tr')[2:-1]:
    tableBody.append(tb.text.split())
```

```python
y=[] #important
for tb in tableBody:
    print(tb)
    print("***********************")
    x=[]
    for el in tb:
        if el != '-':
            ee=float(el.replace(",",""))
            x.append(ee)
        else:
            x.append(el)
    print(x)
    print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
    y.append(x)
```
